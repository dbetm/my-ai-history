## Download dataset of paintings by some authors

```python
classes = {
    "picasso": "Pablo Picasso",
    "vanGogh": "Vincent van Gogh",
    "dali": "Salvador Dal√≠",
    "daVinci": "Leonardo da Vinci",
    "rembrandt": "Rembrandt",
}
```

```python
# copy input to the current working dir -- Util in Kaggle
#!mkdir paintings-by-autor
#!cp -R /kaggle/input/paintings-by-famous-painters/paintings-by-autor/* ./paintings-by-autor
```

```python
path = Path("paintings-by-autor")

if not path.exists():
    path.mkdir()
```

```python
for class_, author in classes.items():
    print(f"Downloading images by {author}")
    dest = (path/class_)
    dest.mkdir(exist_ok=True)
    results = search_images_bing(key, f"Paintings by {author}")
    download_images(dest, urls=results.attrgot("contentUrl"))
    time.sleep(1)
```

```python
fns = get_image_files(path)
fns
```

```python
corrupt_images = verify_images(fns)
corrupt_images
```

```python
# remove failed images
corrupt_images.map(Path.unlink)
```
