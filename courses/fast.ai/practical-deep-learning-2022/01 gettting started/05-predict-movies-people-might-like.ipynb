### MovieLens Dataset

The datasets describe ratings and free-text tagging activities from MovieLens, a movie recommendation service. It contains 20000263 ratings and 465564 tag applications across 27278 movies. These data were created by 138493 users between January 09, 1995 and March 31, 2015.


```python
from fastai.collab import *
from fastai.tabular.all import untar_data, URLs
```

```python
data_path = untar_data(URLs.ML_SAMPLE)

data_loader = CollabDataLoaders.from_csv(csv=data_path/'ratings.csv')

```

```python
learner = collab_learner(data_loader, y_range=(0.5,5.5))
learner.fine_tune(10)
```

This model is predicting movie ratings on a scale of 0.5 to 5.0 to within around 0.6 average error. Since we're predicting a continuous number, rather than a category, we have to tell fastai what range our target has, using the y_range parameter.

```python
learner.show_results()
```
