### Dataset - Adult dataset

It contains some demographic data about individuals (like their education, marital status, race, sex, and whether or not they have an annual income greater than $50k).

```python
from fastai.tabular.all import *
```

```python
data_path = untar_data(URLs.ADULT_SAMPLE)
```

```python
# To get a prettier result with hyperlinks to source code and documentation, install nbdev: pip install nbdev
!pip install nbdev
```

```python
doc(TabularDataLoaders.from_csv)
```

```python
data_loader = TabularDataLoaders.from_csv(
    csv=data_path/"adult.csv",
    path=data_path,
    y_names="salary",
    cat_names=["workclass", "education", 'marital-status', 'occupation', 'relationship', 'race'],
    cont_names=['age', 'fnlwgt', 'education-num'],
    procs=[Categorify, FillMissing, Normalize]
)
```

```python
learner = tabular_learner(data_loader, metrics=accuracy)
```

```python
doc(learner.fit_one_cycle)
```

```python
learner.fit_one_cycle(3)
```

```python
learner.show_results()
```
