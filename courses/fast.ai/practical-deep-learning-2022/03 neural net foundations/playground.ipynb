```python
import pandas as pd
```

```python
dataset_url = "../../../../datasets/mini_iris.csv"

df = pd.read_csv(
    dataset_url,
    header=None,
    names=[
        "sepal_length",
        "sepal_width",
        "petal_length",
        "petal_width",
        "class",
    ]
)
df
```

```python
new_df = df[["sepal_length", "sepal_width", "class"]]
new_df
```

```python
final_df = new_df[new_df["class"].isin(["Iris-setosa", "Iris-versicolor"])]
final_df
```

```python
final_df["class"].unique()
```

```python
final_df.to_csv(dataset_url, index=False)
```
