# Simple Linear Regression

## Importing the libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Importing the dataset

X represents years of experiencie and Y the salary earned

```python
DATASET_PATH = "../../../../datasets/ml_az_course/002_salary_data.csv"
dataset = pd.read_csv(DATASET_PATH)
dataset.head()
```

```python
x = dataset.iloc[:, :-1].values # select values of all rows, for all columns except the last one
y = dataset.iloc[:, -1].values # select values of label, this means, all rows and the last column (salary)
```

```python
x[:10]
```

```python
y[:10]
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X=x_train, y=y_train)
```

## Predicting the Test set results

```python
x_test
```

```python
y_test
```

```python
y_pred = regressor.predict(X=x_test)
y_pred
```

## Visualising the Training set results

```python
plt.scatter(x=x_train, y=y_train, color="red") # plot data-points
plt.plot(x_train, regressor.predict(x_train), color="blue") # trend-line drawn by predictions
plt.title("Salary vs Experience (training set")
plt.xlabel("Year of experience")
plt.ylabel("Salary")
plt.show()
```

## Visualising the Test set results

```python
plt.scatter(x=x_test, y=y_test, color="red") # plot data-points
# trend-line drawn by predictions of the real model, generated by training set
plt.plot(x_train, regressor.predict(x_train), color="blue")
plt.title("Salary vs Experience (test set")
plt.xlabel("Year of experience")
plt.ylabel("Salary")
plt.show()
```
