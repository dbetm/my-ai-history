### Usando merge con datframes dummies

```python
left = pd.DataFrame({
    "Key": ["key1", "key2", "key3"],
    "A": ["a1", "a2", "a3"]
})
left
```

```python
right = pd.DataFrame({
    "Key": ["key3", "key4", "key5"],
    "B": ["b1", "b2", "b3"]
})
right
```

```python
pd.merge(left=left, right=right, on="Key", how="left")
```
