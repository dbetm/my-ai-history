```python
import pandas as pd
```

```python
chicago_data_url = "https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/Traffic_Crashes1.csv"
```

```python
chicago_df = pd.read_csv(chicago_data_url)
chicago_df["CRASH_DATE"] = chicago_df["CRASH_DATE"].apply(
    lambda x: pd.to_datetime(x, errors="coerce", utc=True)
)
```

```python
chicago_df.head(5)
```

```python
chicago_df.groupby(["LIGHTING_CONDITION", "REPORT_TYPE", "CRASH_HOUR"]).agg(
    {"BEAT_OF_OCCURRENCE": "sum"}
)
```

```python
# lograr algo parecido usando pivot table
pd.pivot_table(chicago_df, index=["LIGHTING_CONDITION", "REPORT_TYPE"])
```

### Creando un filtro

```python
chicago_df.filter(["LIGHTING_CONDITION", "REPORT_TYPE"])
```

### Creando un crosstab

```python
# los valores son la incidencia de registros que comparten los valores de condición de luz - tasa de ocurrencia
pd.crosstab(chicago_df["LIGHTING_CONDITION"], chicago_df["BEAT_OF_OCCURRENCE"])
```
