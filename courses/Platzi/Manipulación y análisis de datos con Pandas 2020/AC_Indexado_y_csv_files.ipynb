```python
import numpy as np
import pandas as pd
```

```python
assets_path = "assets"
assets_path_template = "assets/{}"
```

```python
dict_data = {
    'edad': [10, 9, 13, 14, 12, 11, 12],
    'cm': [115, 110, 130, 155, 125, 120, 125],
    'pais': ['co','mx','co','mx','mx','ch','ch'],
    'genero': ['M','F','F','M','M','M','F'],
    'Q1': [5, 10, 8, np.nan, 7, 8, 3],
    'Q2': [7, 9, 9, 8, 8, 8, 9]
}
```

```python
df = pd.DataFrame(dict_data)
df
```

```python
df.to_csv(assets_path_template.format("test.csv"), index=False)
```

```python
!ls {assets_path}
```

### Otros formatos

```python
# Excel, necesario tener instalado "openpyxl"
df.to_excel(assets_path_template.format("test.xlsx"), sheet_name="hoja 1")
pd.read_excel(assets_path_template.format("test.xlsx"))
```

```python
# json
df.to_json(assets_path_template.format("test.json"))
pd.read_json(assets_path_template.format("test.json"))
```

```python
# pickle: para datasets de tama√±o mediano
df.to_pickle(assets_path_template.format("test.pkl"))
pd.read_pickle(assets_path_template.format("test.pkl"))
```

```python
# parquet: formato usado con Apache, necesario instalar fastparquet
df.to_parquet(assets_path_template.format("test.parquet"))
pd.read_parquet(assets_path_template.format("test.parquet"))
```

```python
# hdf: formato usado en Hadoop, necesario instalar pytables
df.to_hdf(assets_path_template.format("test.hdf"), key="data", format="table")
pd.read_hdf(assets_path_template.format("test.parquet"))
```
