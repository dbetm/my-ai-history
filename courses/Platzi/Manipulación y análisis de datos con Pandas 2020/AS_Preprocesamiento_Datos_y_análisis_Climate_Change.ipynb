### Datasets sobre el cambio climático

- Topic: https://data.worldbank.org/topic/climate-change
- CO2 (kt): https://data.worldbank.org/topic/EN.ATM.CO2E.KT
- Forest (km2): https://data.worldbank.org/topic/AG.LND.FRST.K2
- Agricultura (km2): https://data.worldbank.org/topic/AG.LND.AGRI.K2
- Population (7mM): https://data.worldbank.org/topic/SP.POP.TOTL
- Electric power consumation (kWh per capita): https://data.worldbank.org/topic/EG.USE.ELEC.KH.PC

```python
# dataset paths
base_path = "/home/david/ML_DL/datasets/World Bank Climate Change"

agri_path = "{}/API_AG.LND.AGRI.K2_DS2_en_csv_v2_3933252/API_AG.LND.AGRI.K2_DS2_en_csv_v2_3933252.csv".format(base_path)

fore_path = "{}/API_AG.LND.FRST.K2_DS2_en_csv_v2_3932352/API_AG.LND.FRST.K2_DS2_en_csv_v2_3932352.csv".format(base_path)

elec_path = "{}/API_EG.USE.ELEC.KH.PC_DS2_en_csv_v2_3930745/API_EG.USE.ELEC.KH.PC_DS2_en_csv_v2_3930745.csv".format(base_path)

co2e_path = "{}/API_EN.ATM.CO2E.KT_DS2_en_csv_v2_3888754/API_EN.ATM.CO2E.KT_DS2_en_csv_v2_3888754.csv".format(base_path)

popu_path = "{}/API_SP.POP.TOTL_DS2_en_csv_v2_3918567/API_SP.POP.TOTL_DS2_en_csv_v2_3918567.csv".format(base_path)
```

```python
df_agri = pd.read_csv(agri_path, header=2)
df_fore = pd.read_csv(fore_path, header=2)
df_elec = pd.read_csv(elec_path, header=2)
df_co2e = pd.read_csv(co2e_path, header=2)
df_popu = pd.read_csv(popu_path, header=2)
```

```python
df_popu
```

```python
# seleccionar solo algunos datos de los dataframes leídos hasta ahora
cols = ["Country Name", "Country Code"] + list(map(str, range(1971, 2015)))
cols[:5]
```

**Formatear cada dataframe**

```python
# melt -> convertir las columnas de años como parte de filas
def format_df(df, col_name: str):
    return df.loc[:, cols].melt(id_vars=["Country Name", "Country Code"]).rename(
        columns={
            "variable": "date",
            "Country Name": "Country",
            "Country Code": "name",
            "value": col_name
        }
    )
```

```python
df_agri = format_df(df_agri, col_name="agriculture")
df_fore = format_df(df_fore, col_name="forest")
df_elec = format_df(df_elec, col_name="electricprod")
df_co2e = format_df(df_co2e, col_name="co2")
df_popu = format_df(df_popu, col_name="population")
```

```python
df_agri
```

```python
# cambiar el tipo de dato de la columna date
df_popu["date"] = df_popu["date"].astype(float)
df_fore["date"] = df_fore["date"].astype(float)
df_agri["date"] = df_agri["date"].astype(float)
df_elec["date"] = df_elec["date"].astype(float)
df_co2e["date"] = df_co2e["date"].astype(float)
```

**Leer el dataset de temperaturas medias**

```python
temperature_path = (
    "/home/david/ML_DL/datasets/Climate Change: Earth Surface Temperature Data/"
    "GlobalLandTemperaturesByCountry.csv"
)
df_temp = pd.read_csv(temperature_path)
df_temp
```

**Formatearlo**

```python
df_temp["year"] = pd.to_datetime(df_temp["dt"])
filter_idxs = df_temp["year"] > pd.to_datetime("1970-01-01")
df_temp = df_temp[filter_idxs]
```

```python
df_t_avg = df_temp.groupby(
    ["Country", pd.Grouper(key="year", freq="1Y")]
).agg({"AverageTemperature": [np.mean, np.median]})
```

```python
df_t_med = df_t_avg["AverageTemperature"][["median"]].reset_index()
df_t_med["date"] = df_t_med["year"].dt.year
df_t_med.rename(columns={"median": "temperature"}, inplace=True)
```

```python
df_t_med
```

### Juntar los datasets

```python
df_merge = pd.merge(
    df_t_med[["Country", "temperature", "date"]],
    df_popu,
    on=["Country", "date"],
    how="inner"
)
df_merge
```

```python
df_merge = pd.merge(df_merge, df_co2e)
df_merge = pd.merge(df_merge, df_elec)
df_merge = pd.merge(df_merge, df_agri)
df_merge = pd.merge(df_merge, df_fore)
df_merge
```

```python
# eliminar filas con NaNs
df_climate = df_merge.dropna()
```
