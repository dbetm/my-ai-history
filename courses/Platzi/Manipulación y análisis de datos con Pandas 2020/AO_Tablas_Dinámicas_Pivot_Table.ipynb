```python
import numpy as np
import pandas as pd
```

```python
pd.options.display.float_format = "{:,.3f}".format
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df
```

```python
df.groupby(["sex", "time"])[["total_bill"]].mean()
```

```python
df_gp = df.groupby(["sex", "time"])[["total_bill"]].mean().reset_index()
df_gp
```

```python
# reestructurar df_gp
df_gp.pivot_table(values="total_bill", index="sex", columns="time")
```

**Entender tabla pivote**

![](https://pandas-docs.github.io/pandas-docs-travis/_images/reshaping_pivot.png)

```python
# se puede crear la tabla pivote directamente, ya que por defecto la funciÃ³n de agregaciÃ³n es mean
df.pivot_table(values="total_bill", index="sex", columns="time")
```

```python
# se puede modificar dicha funciÃ³n de agregaciÃ³n
df.pivot_table(values="total_bill", index="sex", columns="time", aggfunc=np.max)
```

```python
# con mÃ¡s de una funciÃ³n de agregaciÃ³n
df_p = df.pivot_table(values="total_bill", index="sex", columns="time", aggfunc=[np.max, np.min])
df_p
```

```python
# aplicar unstack, deshacer categorÃ­as -> columnas vuelven a estar en las filas
df_p.unstack().to_frame()
```
