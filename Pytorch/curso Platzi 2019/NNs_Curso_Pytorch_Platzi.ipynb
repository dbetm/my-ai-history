```python
import torch
import torch.nn as nn

linear = nn.Linear(in_features=4096, out_features=10) # Fully connected - tensores 1 dim de input
conv = nn.Conv2d(in_channels=3, out_channels=9, kernel_size=3, stride=1, padding=1) # modifica canales
relu = nn.ReLU(False)
```

```python
x = torch.randn(4096)
y = linear(x)
y.size()
```

```python
x = torch.randn(1, 3, 7, 7)
y = conv(x)
z = relu(x)

print(x)
print('-'*42)
print(y)
print('-'*42)
print(z)
```
